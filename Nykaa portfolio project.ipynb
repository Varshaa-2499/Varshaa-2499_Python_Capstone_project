## **Business Requirement**

#### **KPI's Reqirement**

```python
#Total sales
Total_sales =df['Price_USD'].sum()

#Average sales
Avg_sales =df['Price_USD'].mean()

#no of items sold
No_of_items_sold =df['Category'].count()

#Average ratings
Avg_ratings =df['Rating'].mean()

#output to display
print(f"Total sales: ${Total_sales:,.0f}")
print(f"Average sales: ${Avg_sales:,.1f}")
print(f"Number of items sold: {No_of_items_sold:,.0f}")
print(f"Average ratings: {Avg_ratings:,.0f}")
```

## **Charts Requirements**

#### **Total Sales by main ingredient**

```python
sales_by_Main_Ingredient = df.groupby('Main_Ingredient')['Price_USD'].sum()
plt.pie(sales_by_Main_Ingredient, labels = sales_by_Main_Ingredient.index,
        autopct ='%.1f%%',
        startangle =90)
plt.title('Sales by Main Ingredient')
plt.axis('equal')
plt.show()
```

## **Total sales by Category**

```python
sales_by_category = df.groupby('Category')['Price_USD'].sum().sort_values(ascending= False)
plt.figure(figsize=(13,8))
bars=plt.bar(sales_by_category.index, sales_by_category.values)

plt.xticks(rotation =-90)
plt.xlabel('Category')
plt.ylabel('Total sales')
plt.title('Total Sales by Category')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

## **Total sales by cruelty free products**

```python
grouped = df.groupby(['Category','Cruelty_Free'])['Price_USD'].sum().unstack()

ax = grouped.plot(kind ='bar',figsize =(15,9), title ='Total Sales by cruelty free products based upon category')
plt.xlabel('Category')
plt.ylabel('Total sales')
plt.legend(title ='Cruelty Free')

plt.tight_layout()
plt.show()
```

## **Top 10 Brands by review count**

```python
brand_reviews = df.groupby('Brand')['Number_of_Reviews'].sum().sort_index()
top_10_brands = brand_reviews.sort_values(ascending = False).head(10)

plt.figure(figsize=(12,3))
plt.plot(top_10_brands.index, top_10_brands.values, marker='o', linestyle='-',color='teal')


plt.xlabel('Brand')
plt.ylabel('Total number of reviews')
plt.title('Top 10 Brands by review count')

plt.ticklabel_format(style='plain',axis='y')

plt.grid(True)
plt.tight_layout()
plt.show()
```

## **Bottom 6 Products based on rating**

```python
product_rating = df.groupby('Product_Name')['Rating'].sum().reset_index()
bottom_6_product = product_rating.sort_values('Rating',ascending = False).tail(6)

plt.figure(figsize=(8,4))
ax =sns.barplot(x='Rating',y='Product_Name',data =bottom_6_product, color='red')


plt.xlabel('Rating')
plt.ylabel('Products')
plt.title('Bottom 6 products with low rating')

plt.tight_layout()
plt.show()
```
